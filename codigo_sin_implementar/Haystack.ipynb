```python
#quitar \n
#pner todo en minusculas
#quitar paginas web
import nltk
import re
def limpieza_passages(texto):
  
  texto=texto.replace("\n"," ")
  texto=texto.replace(" - ","")
  texto=texto.lower()
  texto = re.sub(r'^https?:\/\/.*[\r\n]*', '', texto, flags=re.MULTILINE)
 
  return texto
```

```python
def archivos(empresa,año,formato='txt'):
  '''
  Coge los archivos de cada empresa y año que sean de un formato determinado
  Args:
            empresa (str): empresa de la que se sacan los archivos
            año (str): año del que se sacan los archivos
            formato (str): se cogeran los archivos de este formato , por defecto txt

  Return : una lsita con los archivos del formato indicado
  '''
  path_origen='/content/drive/MyDrive/DATO-ingles/'+empresa+'/'+año+'/'

  contenido = os.listdir(path_origen)
  contenido_nombres=[]
  for archivo in contenido:
    tipo=archivo[-3:]
    if tipo ==formato:
      contenido_nombres.append(archivo[:-4])
  
  return contenido_nombres
```

```python
class Buscador:

    def __init__(self, empresa, año):
        self.empresa = empresa
        self.año = año

    def get_path(self):
        """me da el path correspondiente de ese año esa empresa"""
        path_origen='/content/drive/MyDrive/DATO-ingles/'+self.empresa+'/'+str(self.año)+'/'
 
        return path_origen

    def get_documents(self):
        """me da los documentos que hay de esa empresa para su año"""
        path_origen=self.get_path()
  
        contenido = os.listdir(path_origen)
      
        return contenido
    

    def crear_document_store(self):
      host = os.environ.get("ELASTICSEARCH_HOST", "localhost")
      document_store = ElasticsearchDocumentStore(host=host, username="", password="", index="document")
      return document_store
    def archivos_to_docs(self, split_par=True):
      """ convertimos los archivos txt traducidos a el modo determido de haystack """
      docs = convert_files_to_docs(dir_path=self.get_path(), split_paragraphs=split_par)
      return docs

    def clean_docs(self, docs):
      """ limpieza de los docuemntos"""
      preprocessor = PreProcessor(
      clean_empty_lines=True,
      clean_whitespace=True,
      clean_header_footer=False,
      split_by="word",
      split_length=10,
      split_overlap=3, 
      split_respect_sentence_boundary=True,
      language = "en"
      )
      docs_clean = preprocessor.process(docs)
    
      return docs_clean

    def escribe_documentos(self, document_store, docs):
      document_store.write_documents(docs)
    
    def crear_retriever( self, modelo_ret, formato,document_store):
      retriever = EmbeddingRetriever(
        document_store=document_store,
        embedding_model=modelo_ret,
        model_format=formato,
      )
      
      
      return retriever

    def actualizar_embedding(self, retriever, document_store):
      document_store.update_embeddings(retriever)

    def crear_reader(self, modelo_rea):
  
      reader = FARMReader(model_name_or_path=modelo_rea, use_gpu=True)
      return reader

    def crear_pipeline(self, reader, retriever):
      
      pipe = ExtractiveQAPipeline(reader,retriever)
      return pipe

    def preguntar(self,pipeline, query):
  
      prediction = pipeline.run(
        query=query
      )
      return print_answers(prediction)

    def mejor_respuesta(self,pipeline, query):
  
      prediction = pipeline.run(
        query=query
      )
      respuesta=prediction['answers'][0]
      return [respuesta.answer,respuesta.context, respuesta.score]
```

```python
def crear_pipeline(empresa, año):
  # 1º paso:crear la clase
  busc=Buscador(empresa, año)
  # 2º paso:crear docuemnt store
  print('crear doc store')
  doc_store=busc.crear_document_store()
  # 4º paso: convertir archivos txt traducidos a tipo docs (compatible con haystcak)
  print('convertir archivos txt a doc')
  docs=busc.archivos_to_docs()
  # 5º paso:limpio docuemntos
  print('limpiar archivos')
  docs_clean=busc.clean_docs(docs)
  # 6º paso: los mete al docstore
  print('meter archivos a doc store')
  busc.escribe_documentos(docs=docs_clean,document_store=doc_store)
  # 7º paso:crear retriever
  print('crear retriever')
  retriever=busc.crear_retriever(modelo_ret="sentence-transformers/multi-qa-mpnet-base-dot-v1",formato="sentence_transformers", document_store=doc_store)
  busc.actualizar_embedding(retriever=retriever, document_store=doc_store)
  # 8º paso:crear reader
  print('crear reader')
  reader=busc.crear_reader("deepset/roberta-base-squad2")
  # 9º paso:crear pipeline
  pipeline=busc.crear_pipeline(reader=reader,retriever=retriever)

  return [busc, pipeline]
```

```python
def imprimir_respuestas_feo(buscador, pipeline, diccionario_preguntas,dimension, empresa, año):
  nombre='/content/drive/MyDrive/DATO-ingles/'+empresa+'/'+año+'/resultados.txt'
  f = open(nombre, 'w')
  for variable in diccionario_preguntas:
    
    
    f.write('\n')
    f.write('---------------------------------------------')
    f.write('\n')
    f.write('la variable que estamos buscando es: '+ variable)
    f.write('\n')
    query=diccionario_preguntas[variable]
    f.write('la query con la que buscamos : ' + query)
    respuesta=buscador.preguntar(pipeline, query)
    f.write('\n')
    f.write('la respuesta que da el modelo es : ')
    f.write('\n')
    f.write(respuesta)
    
    f.write('\n')
    
    f.write('\n')
    f.write('---------------------------------------------')
    f.write('\n')
  f.close()
```

```python
from haystack.nodes import BM25Retriever,EmbeddingRetriever
# bm25 es un modelo sparse basado en coincidencias con el texto

#retriever = BM25Retriever(document_store=document_store)
#podemos usar retriever mas elevadorados, basados en embedding
# se supone que da mejor resulatdo

retriever = EmbeddingRetriever(
    document_store=document_store,
    embedding_model="sentence-transformers/multi-qa-mpnet-base-dot-v1",
    model_format="sentence_transformers",
)
# Important:
# Now that we initialized the Retriever, we need to call update_embeddings() to iterate over all
# previously indexed documents and update their embedding representation.
# While this can be a time consuming operation (depending on the corpus size), it only needs to be done once.
# At query time, we only need to embed the query and compare it to the existing document embeddings, which is very fast.
document_store.update_embeddings(retriever)
```
